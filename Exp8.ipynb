# Graph Colouring Using Greedy Algorithm

```python
# Loading Libraries
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def greedy_graph_coloring(graph):
  # Initialize the colouring dictionary
  color_assignment = {}
  for node in graph.nodes():
    # Find the colors of adjacent nodes

    adjacent_colors = {color_assignment.get(neighbor) for neighbor in graph.neighbors(node)}
    # Assign the smallest available color
    color = 0
    while color in adjacent_colors:
      color += 1
    color_assignment[node] = color
  return color_assignment
```

```python
def plot_colored_graph(graph, color_assignment):
  pos = nx.spring_layout(graph)
  colors = [color_assignment[node] for node in graph.nodes()]
  nx.draw(graph, pos, with_labels=True, node_color=colors, cmap=plt.cm.rainbow,node_size=700, font_size=14, font_weight='bold',edge_color='black')
  plt.title("Graph coloring using greedy algorithm")
  plt.show()
```

```python
# Example graph creation
G = nx.Graph()
G.add_edges_from([(0, 1), (0, 2), (1, 2), (1, 3),(2, 3), (3, 4)])

# perform Graph Coloring
color_assignment = greedy_graph_coloring(G)
plot_colored_graph(G,color_assignment)
```

# Application problem

C0 and C1
C0 and C2
C1 and C2
C1 and C3
C2 and C3
C3 and C4

The goal is to assign different colour to different courses.
