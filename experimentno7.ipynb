##Matrix Representation of Graphs



UNDIRECTED GRAPH

```python
import numpy as np
# Define the Graph
# Vertices:0,1,2
# Edges with weights: (0-1, w=4), (1-2, w=5), (0-2, w=3)
vertices = 3
edges = [(0,1,4),(1,2,5),(0,2,3)]
# Intialize adjacency matrix
adj_matrix = np.zeros((vertices,vertices), dtype=int)
# Initialize incidence matrix
inc_matrix = np.zeros((vertices,len(edges)), dtype=int)
# Fill matrices
for idx, (u, v, w) in enumerate(edges):
    adj_matrix[u][v] = w

    #Display results
    print("Adjacency Matrix:")
    print(adj_matrix)
    print("\nIncidence Matrix:")
    print(inc_matrix)



```

DIRECTED GRAPH

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
# Step 1: Define edges with weights for a directed grap
# Format: (From_node, to_node, weight)
edges = [(0, 1, 4), (0, 2, 6), (1, 2, 5), (2, 3, 7), (3, 0, 3)]
# Step 2: Create a Directed Graph and Weighted edges
G = nx.DiGraph()
G.add_weighted_edges_from(edges)
# Step 3: Draw the weighted graph
pos = nx.spring_layout(G, seed=42) #Layout
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue',  arrows=True)
edge_labels = {(u, v): w for u, v, w in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Weighted Directed Graph")
plt.show()
# Step 4: adjacency Matrix (Weighted)
n = len(G.nodes)
adj_matrix = np.zeros((n, n), dtype=int)
for u, v, w in edges:
    adj_matrix[u][v] = w

print("Adjacency Matrix:")
print(adj_matrix)
```

Problem statement

A company has four routers (0 to 3) connected in a directed network to route data packets.the bandwidths (in Mbps)are:


0 → 1 :10 Mbps

0 → 2 :20 Mbps

1 → 3 :15 Mbps

2 → 3 :25 Mbps

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

edges = [(0, 1, 10), (0, 2, 20), (1, 3, 15), (2, 3, 25)]

G = nx.DiGraph()
G.add_weighted_edges_from(edges)

pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue',  arrows=True)
edge_labels = {(u, v): w for u, v, w in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Weighted Directed Graph")
plt.show()

n = len(G.nodes)
adj_matrix = np.zeros((n, n), dtype=int)
for u, v, w in edges:
    adj_matrix[u][v] = w

print("Adjacency Matrix:")
print(adj_matrix)
```
